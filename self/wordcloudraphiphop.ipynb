```python
import pandas as pd
import wordcloud
import spacy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def clear_text(text):
    nlp = spacy.load('en')
    doc = nlp(text)
    output = " ".join([token.lemma_ for token in doc if not token.is_stop])
    return output
```

```python
LETTERS = '/home/rlfo/Documents/pessoal/bigdatamusic/modeled/letters.csv'
```

```python
all_data = pd.read_csv(LETTERS,sep=';')
```

```python
all_data.sample(2)
```

```python
all_data = all_data.loc[:, ['text']]
```

```python
all_data.sample(3)
```

```python
all_data['text'] = all_data['text'].apply(clear_text) 
```

```python
wc = wordcloud.WordCloud().generate(' '.join(all_data['text']))
plt.figure(figsize=(15,10))
plt.imshow(wc)
plt.axis('off')
```
